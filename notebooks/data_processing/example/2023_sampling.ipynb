```python
# for weekend and holiday

df_bike_rental = pd.read_csv('../data/bike_rental_history/processed/2023_agg_240729.csv')
df_bike_rental.loc[:,'year'] = df_bike_rental.loc[:,'대여일시'].str[:4]
df_bike_rental = df_bike_rental.loc[df_bike_rental.loc[:,'year'] == '2023']

df_bike_rental.loc[:,'date'] = pd.to_datetime(df_bike_rental.loc[:,'대여일시']).dt.date
df_bike_rental.loc[:,'date'] = df_bike_rental.loc[:,'date'].astype(str)

df_holiday = pd.read_csv('../data/holidays.csv')
df_holiday.loc[:,'date'] = pd.to_datetime(df_holiday.loc[:,'year'].astype(str) +'-'+ df_holiday.loc[:,'month'].astype(str) +'-'+ df_holiday.loc[:,'day'].astype(str)).dt.date
df_holiday.loc[:,'date'] = df_holiday.loc[:,'date'].astype(str)

df_bike_rental.loc[:,'weekday'] = pd.to_datetime(df_bike_rental.loc[:,'대여일시']).dt.weekday
df_bike_rental_weekend = df_bike_rental.loc[np.logical_or(df_bike_rental.loc[:,'weekday'].isin([5,6]), df_bike_rental.loc[:,'date'].isin(df_holiday.loc[:,'date']))].copy()
df_bike_rental_weekend = df_bike_rental.drop(['year', 'weekday'], axis=1)
```
