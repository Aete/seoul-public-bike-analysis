```python
import numpy as np
import pandas as pd
import geopandas as gpd

# Load the data
gdf_landuse_magok = gpd.read_file('../../data/land_use/magok_landuse.geojson')
gdf_landuse_magok = gdf_landuse_magok.dropna(subset=['A17'])
gdf_landuse_magok.head(3)
```

```python
gdf_residential_magok = gdf_landuse_magok[gdf_landuse_magok.loc[:,'A17'].str.startswith('1')]
gdf_residential_magok.loc[:,'landuse'] = 1
gdf_residential_magok = gdf_residential_magok.loc[:,['landuse','geometry']]
gdf_residential_magok.head(3)
```

```python
gdf_commercial_magok = gdf_landuse_magok[gdf_landuse_magok.loc[:,'A17'].str.startswith('2')]
gdf_commercial_magok.loc[:,'landuse'] = 2
gdf_commercial_magok = gdf_commercial_magok.loc[:,['landuse','geometry']]
gdf_commercial_magok.head(3)
```

```python
gpd_magok_landuse = gpd.GeoDataFrame(pd.concat([gdf_residential_magok, gdf_commercial_magok], ignore_index=True))
gpd_magok_landuse.loc[:,'geometry'] = gpd_magok_landuse.loc[:,'geometry'].simplify(0.01, preserve_topology=True)
```

```python
gpd_magok_landuse.to_file('../../data/land_use/magok_landuse_processed.geojson', driver='GeoJSON')
```
